## [Poisson Problem with Robin Boundary Conditions](#toc)<a id="problem"></a>

The following sections describe what is referred to in the literature as the one-dimensional Poisson problem with Robin boundary conditions. This is a classical boundary-value problem of mathematics.

Solve the Poisson model problem. Find $u:[a,b]\subset\Reals\rightarrow\Reals$ such that:
  
\begin{align*}
 -\bigl(-k(x)\, u'\bigr)'(x) + f(x) &= 0 \quad\quad \forall \quad\quad x\in\ ]a,b[, \\
 q_n(a) &= h_\text{a}\,\bigl(u(a)-u_{\text{ref},a}\bigr), \\
 q_n(b) &= h_\text{b}\,\bigl(u(b)-u_{\text{ref},b}\bigr).
\end{align*}
   
This problem is linear and has an analytical solution for given data: diffusion coefficient, $k(x)$, and source bias, $f(x)$. The *heat flux* associated to the quantity $u$, is denoted $q := -k(x)\,u'$, and it is often of interest as a derived quantity. Re-writing the above in terms of the differential operator $\Dcal u := -\bigl(-k(\cdot)\, u'\bigr)'(\cdot)$

\begin{align*}
 \Dcal u(x) + f(x) &= 0 \quad\quad \forall \quad\quad x\in\ ]a,b[, \\
 q_n(a) &= h_\text{a}\,\bigl(u(a)-u_{\text{ref},a}\bigr), \\
 q_n(b) &= h_\text{b}\,\bigl(u(b)-u_{\text{ref},b}\bigr).
\end{align*}

The normal diffusive flux at $x=a$ is $q_n(a) = -(-k\,u'(a)) = h_\text{a}\bigl(u(a)-u_\text{ref}\bigr)$, and $x=b$ is $q_n(b) = -k\,u'(b) = h_\text{b}\bigl(u(b)-u_\text{ref}\bigr)$.

The values of the normal flux are given on the two end points of the domain. This is called *natural* boundary condition and in this particular case, the Robin type.

## [Rayleigh-Ritz Method w/ Robin Boundary Conditions](#toc)<a id="rayleigh-ritz-inhomogeneous"></a>

Find $u^*_N \in V_N(a,b)\subset V(a,b) = \bigl\{ u:[a,b]\subset\Reals\rightarrow\Reals \bigr\}$ such that it minimizes the Poisson energy norm:

\begin{equation*}
  \norm{u-u^*_N}^2_{\Ecal} = \min\limits_{u_N\, \in\, V_N} \norm{u-u_N}^2_{\Ecal},
\end{equation*}

where 
$V_N(a,b) := \bigl\{ u_N = \sum\limits_{i=1}^N\,c_i\,\phi_i, (u'_N,u'_N) < \infty \bigr\}$, $\{\phi_i\mid i=1\ldots,N\}$ is a basis of $V_N(a,b)$. The Poisson energy norm is defined as $\norm{v}_{\Ecal} := (\Dcal v,v) = (v, \Dcal v)$.

The optimum coefficients $\cvec^* := \{c_i\mid i=1,\ldots,N\}$ solve

\begin{equation*}
 \overset{(N\times N)}{\Amtrx}\,\overset{(N\times 1)}{\cvec^*} = \overset{(N\times 1)}\bvec ,
\end{equation*}

where:
  + $A_{i,j} = (k\,\phi'_i,\phi'_j) + h_\text{a}\phi_j(a)\,\phi_i(a) + h_\text{b}\phi_j(b)\,\phi_i(b) $
  + $b_i = (f,\phi_i) + h_\text{a}\,u_{\text{ref},a} \,\phi_i(a) + h_\text{b}\,u_{\text{ref},b}\, \phi_i(b)$

### Data
Solve problem with parameter values:

 > a = 0 cm  <br>
 > b = 5 cm  <br>
 > u_ref_a = 20 C <br>
 > u_ref_b = 40 C <br>
 > h_a = 25 <br>
 > h_b = 25 <br>
 > k = 42 W/(mK) <br>
 > f = 824604 W/m^2

```python
'''SI units'''
from scipy import constants as unit
unit.second = 1.0
unit.meter = 1.0
unit.cm = unit.centi*unit.meter
unit.watt = 1
unit.kelvin = 1
```

```python
'''Domain'''
x_a = 0*unit.cm
x_b = 5*unit.cm

x_bar = (x_a + x_b)/2.0
x_l   = x_b - x_a

u_ref_a = unit.convert_temperature(20, 'C', 'K')
u_ref_b = unit.convert_temperature(40, 'C', 'K')
delta_u_ref = u_ref_b - u_ref_a
u_ref_bar = (u_ref_a + u_ref_b)/2

# Typical free Convection - air, gases and dry vapors : 0.5 - 1000 (W/(m2K))
# Typical forced Convection - air, gases and dry vapors:  10 - 1000 (W/(m2K))
htc = htc_a = htc_b = 1000*unit.watt/unit.meter**2/unit.kelvin 
```

```python
'''Thermal conductivity coefficient function'''
from pyfem1d.toolkit import TargetFunction   

k_cond = 42*unit.watt/unit.meter/unit.kelvin

shape_pts = [(x_a, k_cond), (x_b, k_cond)]

k_cond_func = TargetFunction(shape_pts, type='linear')
  
k_cond_func.plot(n_plot_pts=20, show_shape_pts=True, 
                title='Thermal Conductivity', y_label=r'$k(x)$ [cm]', x_label='x [cm]', 
                x_scale=1/unit.cm)
```

```python
'''Heating power density function'''
from pyfem1d.toolkit import TargetFunction   

heating_pwr_dens = 824604*unit.watt/unit.meter**3

shape_pts = [(x_a, heating_pwr_dens), (x_b, heating_pwr_dens)]

heating_pwr_dens_func = TargetFunction(shape_pts, type='linear')
  
heating_pwr_dens_func.plot(n_plot_pts=20, show_shape_pts=True, 
                title='Heating Power Density', y_label=r'$f(x)$ [W/m$^3$]', x_label='x [cm]', 
                x_scale=1/unit.cm)
```

### Parametric/Solution Insights

```python
'''Thermal number'''
n_t = heating_pwr_dens / 2 / k_cond * (x_a*(x_bar-k_cond/htc) + x_b*(x_bar+k_cond/htc)) / u_ref_bar
print('Thermal number = ', n_t)
```

```python
'''Normalized thermal lift coeff'''
print('Thermal lift [] = ', 1+n_t)
```

```python
'''Gradient number'''
n_g = heating_pwr_dens * x_bar / k_cond / (delta_u_ref / x_l)
print('N_g = ', n_g)
```

```python
'''Biot number'''
biot = htc * x_l / k_cond / 2
print('Biot =', biot)
```

```python
'''Normalized Asymmetry Coeff'''
print('Asymmetry coeff [] = ', biot/(1+biot) + n_g)
```

```python
'''Exact flux'''
q_x_exact = lambda x: heating_pwr_dens*(x-x_bar) - k_cond*delta_u_ref/x_l * biot/(1+biot)
```

```python
'''Hot spot'''
x_star = x_bar + k_cond/heating_pwr_dens * delta_u_ref / x_l * biot / (1+biot)
if x_a <= x_star <= x_b:
    print('Position of hot spot [cm] = ', x_star/unit.cm)
    print('q_x* = ', q_x_exact(x_star))
else:
    print('No hot spot present')
```

### Finite Element Basis Functions

Specify basis functions distribution on the material domain and degree.

```python
'''Build basis functions'''
from pyfem1d.toolkit import LagrangeFEMBasis 

degree = [2]
n_elem = [1]

femlb = LagrangeFEMBasis(degree=degree, n_elem=n_elem, x_partition=[x_a, x_b])
femlb.plot(n_plot_pts=500, n_legend_cols=15)
```

#### [Assembly of $\Amtrx$](#toc)<a id="assemble-A-spatial"></a>
  + $A_{i,j} = (k\,\phi'_i,\phi'_j) + h_\text{a}\phi_j(a)\,\phi_i(a) + h_\text{b}\phi_j(b)\,\phi_i(b) $

```python
'''Build the matrix of coefficients of the linear system'''
a_mtrx = femlb.inner_product_matrix(k_cond_func, "u'", "v'")
#debug this a_mtrx += femlb.boundary_inner_product_matrix('left', lambda x: htc_a, 'u', 'v')
a_mtrx += femlb.boundary_inner_product_matrix('right', lambda x: htc_b, 'u', 'v')  
```

```python
'''Temporary correction'''
for i,phi_i in enumerate(femlb.basis_func_list):
    for j,phi_j in enumerate(femlb.basis_func_list):
        a_mtrx[i,j] += htc_a * phi_j(x_a)[0] * phi_i(x_a)[0] 
```

```python
try:    
    from pyfem1d.toolkit import matrix_rank
except ModuleNotFoundError:
    assert False, 'You need to provide your own code here. Bailing out.'
    
if matrix_rank(a_mtrx, tol=1e-6) == min(a_mtrx.shape):
    print('matrix is full rank.')
else:
    print('matrix is rank deficient.')

import numpy as np
print(np.linalg.matrix_rank(a_mtrx), a_mtrx.shape)
```

#### [Assembly of $\bvec$](#toc)<a id="assemble-b-spatial"></a>
  + $b_i = (f,\phi_i) + h_\text{a}\,u_{\text{ref},a} \,\phi_i(a) + h_\text{b}\,u_{\text{ref},b}\, \phi_i(b)$

```python
'''Build the load vector'''
# something on the left boundary is wrong; need fixing
b_vec = femlb.inner_product_vector(heating_pwr_dens_func, 'v')
#debug this b_vec += femlb.boundary_inner_product_vector('left', lambda x: htc_a * u_ref_a, "v")
b_vec += femlb.boundary_inner_product_vector('right', lambda x: htc_b * u_ref_b, "v")
```

```python
'''Temporary correction'''
for i, phi_i in enumerate(femlb.basis_func_list):
    b_vec[i] += htc_a * u_ref_a * phi_i(x_a)[0]
```

#### [Compute Rayleigh-Ritz Solution](#toc)<a id="assemble-b-spatial"></a>

```python
'''Compute optimal coefficients'''
import numpy as np
c_star_vec = np.linalg.solve(a_mtrx, b_vec)
```

```python
'''Build the best approximation function in V_N'''

def u_star(x):
    u = femlb.evaluation_matrix(x)@c_star_vec
    return u[0] if len(u) == 1 else u
```

```python
'''Build the derivative '''

def u_prime_star(x):
    u = femlb.evaluation_matrix(x, derivative=True)@c_star_vec
    return u[0] if len(u) == 1 else u
```

```python
'''Build the heat flux of the best approximation in V_N'''

def heat_flux_x(x):
    q = - k_cond_func(x) * u_prime_star(x)
    return q[0] if len(q) == 1 else q
```
